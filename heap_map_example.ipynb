<center><h1>Seaborn</h1></center>

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Basic Example</h2>

```python
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])
print(df)
```

```python
p1 = sns.heatmap(df)
```

<h2> Advanced Exmaple </h2>

```python
# the data, here we use a random 24 by 10 matrix
data = np.random.randn(24, 10)

# the x axis
days = ['01/07','02/07','03/07','04/07','05/07','06/07','07/07','08/07','09/07','10/07']

# transform to dataframe
df = pd.DataFrame(data, columns = days)

# plot
ax = plt.axes()
sns.heatmap(df, xticklabels=2, yticklabels=2, ax = ax)

ax.set_title('Title here')
ax.set_xlabel("x label")
ax.set_ylabel("y label")

plt.show()
```
